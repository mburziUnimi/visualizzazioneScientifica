```python
df =  pd.read_csv('database/usa-gender-21.csv')
gender = df["Gender"]
val = df["V"]
colors = ["#D7191C", "#2C7BB6"] 
plt.figure(facecolor='#FFFAF6')
plt.pie(val, labels=gender, colors=colors,
autopct='%1.1f%%', startangle=270)
plt.title("2021")
plt.show()
```
