```python
import pandas as pd
from matplotlib import pyplot as plt
#sources<br>
#https://www.formula1.com/en/latest/article.f1-race-weekends-attended-by-over-four-million-fans-in-2018.lILWTUMmCkWAAoGE4aYGo.html<br>
#https://en.wikipedia.org/wiki/2019_United_States_Grand_Prix<br>
#https://en.wikipedia.org/wiki/2021_United_States_Grand_Prix<br>
#https://en.wikipedia.org/wiki/2022_United_States_Grand_Prix<br>
#https://f1destinations.com/2023-united-states-grand-prix-attendance/#:~:text=432%2C000%20Attend%202023%20United%20States%20Grand%20Prix%20at%20COTA,-October%2022%2C%202023
```

```python
df = pd.read_csv('database/cota-attendances.csv')
plt.title("Presenze a Circuit Of The Americas")
plt.xlabel("Anno")
plt.ylabel("Presenze (in migliaia)")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.plot(df.Year, df.Attendance, 'b.-', color='#ABD9E9')
plt.gca().set_facecolor('#FFFAF6')
plt.gcf().set_facecolor('#FFFAF6')
plt.show()
```
