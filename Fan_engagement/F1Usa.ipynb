```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
from datetime import datetime
#sources
#https://trends.google.com
```

```python
fonte = pd.read_csv('database/f1-us.csv')
dates = fonte['Date']
newdates = []
for i in dates:
    date = datetime.strptime(str(i), '%Y-%m-%d')
    newdates.append(date)
fonte['Date2'] = newdates
plt.figure(figsize=(10,8))
plt.title("Popolarit della ricerca 'Formula 1' negli Stati Uniti",fontdict={'fontsize':15,'fontweight':'bold'})
plt.xlabel('Anno')
plt.ylabel('Popolarit')
plt.plot(fonte['Date2'], fonte['V'], color='#2C7BB6')
plt.gca().set_facecolor('#FFFAF6')
plt.gcf().set_facecolor('#FFFAF6')
ax = plt.gca()
ax.axvline(datetime(int(2019), int(3), int(8)), color = '#D7191C', linestyle = '--', alpha = 0.5, label='Stagione 1')
ax.axvline(datetime(int(2020), int(2), int(28)), color = '#D7191C', linestyle = '--', alpha = 0.5, label='Stagione 2')
ax.axvline(datetime(int(2021), int(3), int(19)), color = '#D7191C', linestyle = '--', alpha = 0.5, label='Stagione 3')
ax.axvline(datetime(int(2022), int(3), int(11)), color = '#D7191C', linestyle = '--', alpha = 0.5, label='Stagione 4')
ax.axvline(datetime(int(2023), int(2), int(24)), color = '#D7191C', linestyle = '--', alpha = 0.5, label='Stagione 5')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', facecolor = '#FFFAF6')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.gcf().autofmt_xdate()
```
