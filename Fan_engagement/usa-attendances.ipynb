```python
import matplotlib.pyplot as plt
import pandas as pd
#sources<br>
#https://f1destinations.com/ranked-the-top-attended-formula-1-races-in-2023/
```

```python
df =  pd.read_csv('usa-attendances.csv')
gp = df["GP"]
att = df["Attendance"]
colors = ["#D7191C", "#FDAE61", "#ABD9E9", "#2C7BB6"] 
plt.figure(facecolor='#FFFAF6')
plt.pie(att, labels=gp, colors=colors,
autopct='%1.1f%%', startangle=270)
```

```python
plt.show()
# %%
```
