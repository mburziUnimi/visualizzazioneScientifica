```python
import matplotlib.pyplot as plt
import pandas as pd
#sources<br>
#https://espnpressroom.com/us/press-releases/2022/11/formula-1-season-smashes-u-s-television-viewership-records/#:~:text=Persons%20ages%2018%2D34%20averaged,percent%20of%20the%202022%20audience
```

```python
df =  pd.read_csv('usa-age-split-21.csv')
age = df["Age"]
val = df["V"]
colors = ["#d7191c","#fdae61","#ffffbf", "#abd9e9", "#2c7bb6"] 
plt.figure(facecolor='#FFFAF6')
plt.pie(val, labels=age, colors=colors,
autopct='%1.1f%%', startangle=270)
plt.title("2021")
plt.show()
# %%
```
