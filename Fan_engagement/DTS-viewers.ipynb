```python
import pandas as pd
from matplotlib import pyplot as plt
#sources
#https://theathletic.com/4402239/2023/04/13/f1-formula-one-drive-to-survive-ratings/
df = pd.read_csv('database/DTS-viewers.csv')
```

```python
plt.xlabel("Stagione e anno di distribuzione")
plt.ylabel("Spettatori (in migliaia)")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.ticklabel_format(axis='y', style='sci', scilimits=(3,3))
plt.plot(df.Date, df.V, 'b.-', color = '#2C7BB6')
plt.gca().set_facecolor('#FFFAF6')
plt.gcf().set_facecolor('#FFFAF6')
plt.show()
```
