```python
import pandas as pd
from matplotlib import pyplot as plt
#sources<br>
#https://f1destinations.com/ranked-how-much-do-f1-tickets-cost-in-2023/
```

```python
data = pd.read_csv("gp-prices-23.csv")
df = pd.DataFrame(data)
```

```python
grandprix = df['GP']
price = df['Price']
colors = ['#ABD9E9'] * len(grandprix)
```

```python
colors[9] = '#D7191C'
colors[12] = '#D7191C'
colors[13] = '#D7191C'
```

```python
plt.bar(grandprix, price, color=colors)
plt.gca().set_facecolor('#FFFAF6')
plt.gcf().set_facecolor('#FFFAF6')
plt.xticks(rotation=30, ha='right')
plt.xlabel('Grand Prix')
plt.ylabel('Prezzo in USD')
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
plt.show()
```
